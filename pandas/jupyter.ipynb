```python
df = pd.read_csv ('../data/data.csv')
df.head(10)
```

```python
df.info()
# datasetnya banyak yang missing value
# bisa dilakukan dropping tapi konsekuensinya sebagian besar data hilang
# fill diisi  tapi tidak objektif,  entah dengan value random, atau dengan mean, median, dan mode. 
```

```python
# praktekin semua saja 
# 1. drop mv
df_drop = df.dropna() 
df_drop.info()
```

```python
# gila data yang dari 700 an ke 40 an doang
```

```python

# 2. fill dengan mean
mn = df['UKT-4'].mean()
df_fillna=df.fillna({"UKT-4": mn})
df_fillna.info()
```

```python
# terlihat data ukt-4 sudah sama dengan data diatasnya yaitu 702 yang sebelumnya 699. Di atribut ukt-4 itu diisi mean dari atribut ukt-4 itu sendiri
```

```python
md = df_fillna['UKT-5'].median()
df_fillna = df.fillna({"UKT-5":md})
df_fillna.info()
```

```python
mode = df['UKT-6'].mode()[0]
df_fillna = df.fillna({'UKT-6':mode})
df_fillna.info()
```
