```python

# ndim, shape, size, and dtype 
# 1. ndim (number of dimentions): berapa jumlah dimensinya atau ada berapa axis?
#   Analogi:
#   --ndim = 1: Seperti garis lurus (hanya punya Panjang).
#   --ndim = 2: Seperti kertas/tabel (punya Panjang & Lebar).
#   --ndim = 3: Seperti kubus (punya Panjang, Lebar, & Tinggi).
#   Di Teks: a.ndim hasilnya 2. Artinya array a adalah sebuah tabel datar (2 Dimensi).

a.ndim
```

```python
# 2. shape (bentuk, ukuran rinci): berapa panjang masing-masing sisi dimensinya?
#    Analogi: Jika ndim memberi tahu itu adalah "Tabel", maka shape memberi tahu "Tabel ukuran berapa?".
#    Di Teks: a.shape hasilnya (3, 5).
#   --Angka 3 = Ada 3 Baris.
#   --Angka 4 = Ada 4 Kolom.
#   Hubungan: Jumlah angka di dalam (3, 5) ada dua buah, ini membuktikan ndim-nya adalah 2.

a.shape
```

```python
# 3. size (Total Ukuran/Isi): Ada berapa total elemen (angka) di dalam seluruh array ini?
#   Rumus: Hasil perkalian dari shape.
#   Di Teks: a.size hasilnya 15.
#   Didapat dari: 3 baris dikali 5 kolom = 15 kotak data.
#   Analogi: Jika shape adalah ukuran gedung (3 lantai x 4 kamar), maka size adalah total jumlah kamar di gedung itu (12 kamar).

a.size
```

3. np.empty() 
Maksudnya "Empty" (Kosong) Itu Apa?
Di dunia komputer, "Empty" di sini bukan berarti "Hampa" atau "Nol". Artinya adalah: "Tolong pesankan tempat di memori, tapi jangan dibersihkan isinya." Ketika komputer memberikan tempat memori untuk array tersebut, komputer tidak menghapus jejak data yang pernah ada di sana sebelumnya. Jadi, angka yang muncul apapun itu adalah "Sampah Memori" (Garbage Value)â€”sisa-sisa data dari program lain yang kebetulan pernah memakai alamat memori tersebut.
Analogi: Kamar Hotel
Bayangkan Anda adalah resepsionis hotel.
np.zeros() (Array isi Nol): Tamu datang. Sebelum Anda kasih kunci, Anda suruh Cleaning Service bersihkan kamar sampai kinclong, sprei diganti baru, lantai disapu bersih (di-set jadi 0).
Hasil: Kamar bersih (nilainya 0 semua). Konsekuensi: Agak Lambat, karena butuh waktu untuk bersih-bersih.

np.empty() (Array Kosong): Tamu datang. Anda langsung lempar kuncinya. "Nih kamarnya, langsung masuk aja!". Anda tidak menyuruh Cleaning Service masuk dulu.
Hasil: Di dalam kamar mungkin masih ada sisa sampah tamu sebelumnya (bisa ada botol minum, kertas bekas, dll). Itulah angka acak 3.14 atau 42 tadi. Konsekuensi: Sangat Cepat, karena tidak perlu waktu bersih-bersih.

kenapa pakai empty? karena speed, juga biasa digunakan sebagai wadah sementara, sehingga ngapain memori dibersihkan jadi 0 kalau satu detik kemudian ditimpa juga.

```python
np.empty(2)
```

5. np.linspace()
Ini adalah fungsi penyelamat ketika malas menghitung matematika. np.linspace adalah singkatan dari Linear Space (Spasi yang Linear/Rata). bedanya dengan metode biasa, menggunakan analogi Memasang Tiang Lampu.
Analogi: Proyek Tiang Lampu Jalan
Bayangkan Anda adalah kontraktor jalan raya sepanjang 10 meter (dari meter ke-0 sampai meter ke-10).

    1. Cara Lama (np.arange): Fokus pada Langkah Bos Anda bilang: "Pasang tiang dari meter 0 sampai 10, jarak antar tiang harus 2.5 meter." Anda harus menghitung manual: 0, lalu 2.5, lalu 5... dst. Anda mengatur Jaraknya. Anda tidak tahu nanti totalnya jadi berapa tiang, pokoknya pasang saja.

    2. Cara np.linspace: Fokus pada Jumlah (Target) Bos Anda bilang: "Saya tidak peduli jaraknya berapa, pokoknya dari meter 0 sampai 10, saya mau di situ tertancap tegak 5 tiang dengan jarak yang rata." Di sinilah np.linspace bekerja. Komputer yang akan pusing menghitung jaraknya, Anda tinggal terima beres.


```python
np.linspace(0,10, num=5)


# Bedah Kode

# np.linspace(0, 10, num=5)

# 0 (Start): Mulai dari mana? (Titik Awal).
# 10 (Stop): Berhenti di mana? (Titik Akhir). Penting: Di linspace, angka 10 ini ikut masuk (inklusif).
# num=5 (Number): Mau dipotong jadi berapa bagian/titik?
# Hasilnya: array([ 0. , 2.5, 5. , 7.5, 10. ])

# Komputer secara otomatis membagi jarak 0 ke 10 agar muat persis untuk 5 angka. (0 -> 2.5 -> 5.0 -> 7.5 -> 10.0)
# Kapan Pakai Ini?
# Fungsi ini sangat sering dipakai saat Membuat Grafik (Plotting).
# Misalnya, Anda ingin menggambar kurva mulus. Anda butuh ribuan titik koordinat dari X=0 sampai X=100. Daripada bikin manual, cukup ketik: x = np.linspace(0, 100, num=1000) Maka Anda akan punya 1000 titik yang tersebar rapi untuk menggambar garis kurva yang halus.
```

```python
# manual override, secara default tipe data array dalam numpy adalah float64 (presisi tinggi, aman untuk perhitungan matematika, cirinya ada titik di belakang angak, 0., 1., ). kenapa dan kapan harus pakai dtype? untuk hemat ram, data gambar (biasanya format uint8(bilangan bulat 0-255), bukan desimal), index/id (no. antrian, no. id jarang pakai desimal bahkan tidak ada)

# dg menetapkan langsung tipe data
np.ones(4, dtype=np.int64)
```
